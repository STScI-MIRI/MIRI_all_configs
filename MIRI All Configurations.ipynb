# READ ME

The intent of this notebook is to generate a list of every possible configuration of MIRI observations, and check each configuration in CRDS for a reference file. Possible configurations are informed by https://innerspace.stsci.edu/pages/viewpage.action?spaceKey=JWST&title=MIRI+Allowed+Configurations

DISCLAIMER: This notebook is NOT intended for frequent use and is therefore not designed to be abstract or generalized. It is also not intended to be updated regularly. Should its use be required beyond the original design it will need to be re-evaluated for corrections to each MIRI configuration. 

```python
import pandas as pd
df = pd.DataFrame(columns=['Exposure Type', 
                           'Readout Mode',
                           'Filter',
                           'Band',
                           'Channel',
                           'Detector',
                           'SubArray'])
pd.set_option('display.max_rows', None)
```

# Imaging

## MIR_IMAGE

```python
exp_type = ['MIR_IMAGE']
mode = ['FAST', 'SLOW', 'FASTGRPAVG']
filt = ['F560W', 
        'F770W', 
        'F1000W',
        'F1130W', 
        'F1280W', 
        'F1500W',
        'F1800W', 
        'F2100W', 
        'F2550W',
        'FLENS',
        'OPAQUE',
        'FND',
        'F1065C',
        'F1140C',
        'F1550C',
        'F2300C',
        'P750L']
sub = ['FULL',
       'BRIGHTSKY',
       'SUB256',
       'SUB128',
       'SUB64',
       'SLITLESSPRIM']
```

```python
# MIR_IMAGE
for m in mode:
    for f in filt:
        for s in sub:
            if m == 'SLOW' and s != 'FULL':
                pass
            else:
                temp = {'Exposure Type':'MIR_IMAGE',
                        'Readout Mode':m, 
                        'Filter':f, 
                        'SubArray':s}
                df = df.append(temp, ignore_index=True)
```

# Check CRDS

```python
import crds
```

```python
for i in df.index:
    e = df['Exposure Type'][i]
    
    if 'MRS' in e:
        r = df['Readout Mode'][i]
        b = df['Band'][i]
        c = df['Channel'][i]
        d = df['Detector'][i]
        s = df['SubArray'][i]
        
        params = {'INSTRUME': 'MIRI',
                 'EXP_TYPE': e,
                 'READPATT': r,
                 'BAND': b,
                 'CHANNEL': c,
                 'DETECTOR': d,
                 'SUBARRAY': s}
    else:
        r = df['Readout Mode'][i]
        f = df['Filter'][i]
        s = df['SubArray'][i]
    
        params = {'INSTRUME': 'MIRI',
                 'EXP_TYPE': e,
                 'READPATT': r,
                 'FILTER': f,
                 'SUBARRAY': s}

    result = crds.getrecommendations(parameters=params)
    
    print('Configuration:')
    for key, val in params.items():
        print('{}: {}'.format(key, val))
    
    print('\nResult: ')
    for key, val in result.items():
        print('{}: {}'.format(key, val))
        
    print('\n')
```
